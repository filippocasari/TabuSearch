## Bayesian Optimization

```python
from AI2022MA.DeJongFunctions import DeJong
from bayes_opt import BayesianOptimization
import numpy as np

example = DeJong(5)

def f_to_opt(x, y):
  return - example.evaluate(np.array([[x, y]]))[0]

optimizer = BayesianOptimization(f=f_to_opt,
                                 pbounds={"x": example.range, "y": example.range}, 
                                 random_state=123)
```

```python
optimizer.maximize(init_points=200, n_iter=500,)
print(optimizer.max)
```
